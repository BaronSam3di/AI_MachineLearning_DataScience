### General Principles

- **Consistency**: All objects (basic or composite) share a consistent interface composed of a limited set of methods. This interface is documented in a consistent manner for all objects.

- **Inspection**: Constructor parameters and parameter values determined by learning algorithms are stored and exposed as public attributes. 

- **Non-proliferation of classes**: Learning algorithms are the only objects to be represented using custom classes. Datasets are represented as NumPy arrays or SciPy sparse matrices,,  Hyper-parameters names and values are present as standard Python strings or numbers whenever possible. This keeps scikitlearn easy to combine with other libraries. 

- **Composition**: Many Machine Learning tasks re expressible as sequences or combination or transformation to data, Some learning algorithms are also naturally viewed as meta-algorithms parameterized on other algorithms. Whenever feasible, such algorithms are implemented and composed from existing building blocks.

- **Sensible defaults** : Whenever an operation requires a user-defined parameter, an appropriate default value is defined by the library. the default value should cause the operation to be performed in a sensible way (giving a baseline solution for the task at hand.)

### Basic steps of using Scikit-Learn API

1. Choose a class of model , appropriate for your data based on initial exploratory analysis.
2. Choose model hyperparameters
3. Arrange data in features matrix and target array. 
4. Fit model to data 
5. Apply trained model to new data after the training.

## Supervised Learning example: Simple Linear Regression
### "When x is _, y is _"
Linear Regression aims to establish if there is a **statistically significant** relationship between two variables. Eg Income and spending, location and price, sugar and health. We may also want to **forcast** new observations. IT is called Linear because if we plot it on a bi-dimensional plot it will create a straight line.

- Dependent variable: the value we want to forecast/explain. Denoted as `Y`.
- Independent variable: the value that explains Dependent. Denoted as `X`.
- Errors/Difference between the real data points and the Line of the regression model represented as $\epsilon$ 

The Linear equation model: $y$ = $\beta_0$ + $\beta_1$ $x$ + $\epsilon$

Where 
- $\beta_0$ the constant/intercept of x
- $\beta_1$ the coeficient/slope of x
- $\epsilon$ is the error term we are trying to minimise

[More Theory on Youtube](https://www.youtube.com/watch?v=owI7zxCqNY0)

```python
x = 10 * np.random.rand(100)      # 100 random numbers for our features matrix , multiplied by 10
```

```python
# Target array to see if our model will learn this structure
y = 3 * x  + np.random.randn(100)    # randn is normal distributon
```

```python
plt.scatter(x,y); # "when x is _ , y is _"
```

```python
# Step 1. Choose a class of model - (this will be the overal model, in this example Linear Regression)
from sklearn.linear_model import LinearRegression
```

```python
# Step 2. Choose model hyperparameters (The addtionoal attributes of the model, eg SVM:cost, normalized, pre-processed)
MyModel = LinearRegression(fit_intercept=True, copy_X=True, n_jobs=1, normalize=False)
```

```python
# Step 3. Arrange data in features matrix
X = x.reshape(-1, 1) # 
X.shape
```

```python
# Step 4. Fit model to data
MyModel.fit(X,y)
```

```python
# Before we make a prediction, it is interesting to see that this coeficianet is close to the 3 (Constant ) of out data, so it is being noticed.
MyModel.coef_
```

```python
MyModel.intercept_
```

### ...We Now perform the prediction with out Linear Regression Model

```python
x_fit = np.linspace(-1, 11) # Returns evenly spaced numbers between -1 and 11.
```

```python
X_fit = x_fit.reshape(-1,1)
```

```python
y_fit = MyModel.predict(X_fit) # make predictions with the data using the model
```

```python
# we can se the model recovered the orginal features. When x is 10, y is 30; so when x is 1, y is approximatly 3
plt.scatter(x,y);
plt.plot(x_fit,y_fit);
```
